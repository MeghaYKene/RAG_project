```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader(file_path)
pages = []
async for page in loader.alazy_load():
    pages.append(page)
```

```python
import os
os.environ["TOKENIZERS_PARALLELISM"] = "false"
```

```python
#setting embedding model from hugging face 

from langchain_huggingface import HuggingFaceEmbeddings

embed_model = HuggingFaceEmbeddings(model_name="BAAI/bge-base-en-v1.5")
```

```python
#perform semantic chunking (the process starts from here)

#function for semantic chunker 
from langchain_experimental.text_splitter import SemanticChunker

text_splitter = SemanticChunker(
    embed_model,breakpoint_threshold_type = "percentile"
)
```

```python
semantic_chunks = text_splitter.create_documents([d.page_content for d in pages])
```

```python
print(len(semantic_chunks))
```

```python
#process for pineconen vector db 
```

```python
from langchain_core.documents import Document 

documents =[
    Document(page_content=doc.page_content,
            metadata={"chunk_index": idx, "source": "pages"}
    )
    for idx, doc in enumerate(semantic_chunks)
]
```

```python
#generate unique ids
from uuid import uuid4
uuids = [str(uuid4()) for _ in documents]
```

```python
#enter API key to pinecone vector db 
import getpass
import os
import time

from pinecone import Pinecone, ServerlessSpec

if not os.getenv("PINECONE_API_KEY"):
    os.environ["PINECONE_API_KEY"] = getpass.getpass("Enter your Pinecone API key: ")

pinecone_api_key = os.environ.get("PINECONE_API_KEY")

pc = Pinecone(api_key=pinecone_api_key)
```

```python
#creating new index for pinecone
import time

index_name = "semantic-chunk-vectordb01"  # change if desired

existing_indexes = [index_info["name"] for index_info in pc.list_indexes()]

if index_name not in existing_indexes:
    pc.create_index(
        name=index_name,
        dimension=768,
        metric="cosine",
        spec=ServerlessSpec(cloud="aws", region="us-east-1"),
    )
    while not pc.describe_index(index_name).status["ready"]:
        time.sleep(1)

index = pc.Index(index_name)
```

```python
#defining vector_store
from langchain_pinecone import PineconeVectorStore

vector_store = PineconeVectorStore(index=index,embedding=embed_model)
```

```python
vector_store.add_documents(documents=documents,ids=uuids)
```
